```python
def get_valid_name(prompt):
    while True:
        name = input(prompt).strip()   # Remove leading/trailing spaces

        # Check if input is empty
        if not name:
            print("âŒ Name cannot be empty. Please try again.")
            continue

        # Check if input contains only alphabets
        if not name.isalpha():
            print("âŒ Name must contain only letters. Please try again.")
            continue

        return name.upper()   # Convert to uppercase for formatting


# Greeting message
greeting = "âœ¨ Warm Greetings "

# Get validated inputs
first_name = get_valid_name("Enter First Name: ")
last_name = get_valid_name("Enter Last Name: ")

# Display output
print("\n" + greeting + first_name + " " + last_name)
```

```python
class Student:
  def __init__(self,name,age,subjects):
    self.name=name
    self.age=age
    self.subjects=subjects
    self.grades=[]
  def set_grades(self):
    for i in range(self.subjects):
      grade=int(input(f"Enter grade for subject {i+1}: "))
      self.grades.append(grade)
  def get_average(self):
    avg=sum(self.grades)/len(self.grades)
    return avg

Student1=Student("John",20,3)
Student1.set_grades()
print("Average of Student1 is: ",Student1.get_average())
```

```python
class BankAccount:
  def __init__(self,name,balance):
    self.name=name
    self.balance=balance
  def deposit(self,amt):
    print("Rs:",amt,"deposited successfully")
    self.balance+=amt
  def withdraw(self,amt):
    if self.balance>=amt:
      print("Rs:",amt,"withdrawn successfully")
      self.balance-=amt
    else:
      print("Insufficient balance")
  def getBalance(self):
    return self.balance


b1=BankAccount("John",1000)
b1.deposit(500)
b1.withdraw(000)
b1.getBalance()
```

You are given a list of students where each student is represented as a dictionary:

students = [
    {"id": 1, "name": "Ava", "marks": (78, 85, 90)},
    {"id": 2, "name": "Liam", "marks": (88, 79, 92)},
    {"id": 3, "name": "Noah", "marks": (60, 70, 65)}
]


Tasks:

Use loops to calculate the average marks of each student.

Store students with average â‰¥ 80 in a set.

Use if-else to label each student as "Excellent", "Good", or "Needs Improvement".

Print results in a readable format.

```python
students = [ {"id": 1, "name": "Ava", "marks": (78, 85, 90)}, {"id": 2, "name": "Liam", "marks": (88, 79, 92)}, {"id": 3, "name": "Noah", "marks": (60, 70, 65)} ]
label=set({})
for student in students:
  average=sum(student["marks"])/len(student["marks"])
  if average>=80:
    label.add(student['name'])

  if average>=90:
    print(student['name'],"Excellent")
  elif average>=80:
    print(student['name'],"Good")
  else:
    print(student['name'],"Needs Improvement")
print("Above 80 Average Set:",label)
```
